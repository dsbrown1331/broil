```python
from numpy import genfromtxt
import numpy as np
run_v_states_mr = genfromtxt('./results/stress_test/machine_replace_states.csv', delimiter=',')
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
num_states = [10,50,100,200, 500,1000,2000,5000]
run_mean = np.mean(run_v_states_mr, axis=0)
error = np.std(run_v_states_mr, axis=0)
plt.figure(1)
plt.plot(num_states, run_mean,lw=2)
plt.fill_between(num_states, run_mean - error, run_mean + error, alpha=0.5)
plt.ylabel('Run time (s)',fontsize=20)
plt.xlabel("Number of states",fontsize=20)
plt.xticks(fontsize=18) 
plt.yticks(fontsize=18)
#plt.legend(loc='lower right',fontsize=19)
plt.tight_layout()
plt.savefig("./figs/machine_replacement/machine_replacement_states.pdf")
```

```python
run_v_rews_mr = genfromtxt('./results/stress_test/machine_replace_rewards.csv', delimiter=',')

num_rewards = [10,50,100,200, 500,1000,2000,5000]
run_mean = np.mean(run_v_rews_mr, axis=0)
error = np.std(run_v_rews_mr, axis=0)
plt.figure(2)
plt.plot(num_rewards, run_mean,lw=2)
plt.fill_between(num_rewards, run_mean - error, run_mean + error, alpha=0.5)
plt.ylabel('Run time (s)',fontsize=20)
plt.xlabel("Number of rewards in posterior",fontsize=20)
plt.xticks(fontsize=18) 
plt.yticks(fontsize=18)
#plt.legend(loc='lower right',fontsize=19)
plt.tight_layout()
plt.savefig("./figs/machine_replacement/machine_replacement_rewards.pdf")
```
